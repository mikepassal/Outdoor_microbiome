```python
def run_go_term_analysis(gene_list, background_genes, go_term_matrix, go_term_names):
    cont_tables_large = Go_tools.generate_contigency_tables(
        go_annotations=go_term_matrix,
        gene_list=gene_list,
        use_background_genes=True,
        background_genes=background_genes,
    )

    fishers_results_large = Go_tools.fishers_exact_on_contigency_tables(
        all_go_contigency_tables=cont_tables_large,
        original_GO_term_panda=go_term_matrix,
    )
    corrected_fishers_results_large = Go_tools.multi_hypothesis_correct_fishers_exact(
        fishers_results_large
    )
    named_large_fishers_results = corrected_fishers_results_large.merge(
        go_term_names, left_index=True, right_index=True, how="left"
    )
    return named_large_fishers_results
```

```python
cont_tables_large = Go_tools.generate_contigency_tables(
    go_annotations=go_term_matrix,
    gene_list=circadian_1_gene_list_large,
    use_background_genes=True,
    background_genes=background_genes_list,
)

fishers_results_large = Go_tools.fishers_exact_on_contigency_tables(
    all_go_contigency_tables=cont_tables_large,
    original_GO_term_panda=go_term_matrix,
)

corrected_fishers_results_large = Go_tools.multi_hypothesis_correct_fishers_exact(
    fishers_results_large
)

named_large_fishers_results = corrected_fishers_results_large.merge(
    go_term_names, left_index=True, right_index=True, how="left"
)
named_large_fishers_results.sort_values("P_value").head(60)
```

```python
filtered_circadian_1_aligned = filtered_circadian_1.set_index("test")
filtered_circadian_2_aligned = filtered_circadian_2.set_index("test")
filtered_circadian_1_aligned
```

```python
### Smaller is better so if answer is positive then experiment 2 is better, if negative then experiment 1 is better.
### The biggest gain possible is 1
circadian_delta_p = (
    filtered_circadian_1_aligned["p"] - filtered_circadian_2_aligned["p"]
)
circadian_delta_p = circadian_delta_p.to_frame(name="delta_p")
circadian_delta_p = circadian_delta_p.sort_values("delta_p", ascending=False)
circadian_delta_p
```
